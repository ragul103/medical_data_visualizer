```python
import pandas as pd
df = pd.read_csv("medical_examination.csv")
```

```python
df.head()
```

```python
df['overweight'] = (df['weight'] / ((df['height'] / 100) ** 2) > 25).astype(int)
```

```python
df['cholesterol'] = (df['cholesterol'] > 1).astype(int)
```

```python
df['gluc'] = (df['gluc'] > 1).astype(int)
```

```python
df_cat = pd.melt(
    df,
    id_vars=['cardio'],
    value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']
)
```

```python
df_cat = (
    df_cat
    .groupby(['cardio', 'variable', 'value'])
    .size()
    .reset_index(name='total')
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
catplot = sns.catplot(
    data=df_cat, x='variable', y='total',
    hue='value', col='cardio', kind='bar'
)
fig = catplot.fig
```

```python
df_heat = df[df['ap_lo'] <= df['ap_hi']]
df_heat = df_heat[df_heat['height'].between(df['height'].quantile(0.025), df['height'].quantile(0.975))]
df_heat = df_heat[df_heat['weight'].between(df['weight'].quantile(0.025), df['weight'].quantile(0.975))]
```

```python
corr = df_heat.corr()
corr
```

```python
import numpy as np
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, mask=mask, annot=True, fmt=".1f", center=0, square=True, linewidths=.5, cbar_kws={'shrink':0.5})
```
